# Random forest 

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.metrics import accuracy_score
```

```python
para={'n_estimators': range(2,100,2), 'max_depth':range(2,100,2) }
# nbre d'arbre 
# profondeur 
# nbre d'échantillon /arbre 
```

## Début  GridSearch varier les paramètres 

```python
gis_search= GridSearchCV(param_grid=para, estimator=RandomForestClassifier(), scoring='accuracy')
# score : accuracy score
```

```python
# Partie relative à  Gridsearch
```

```python
# splitter les données en données Test et données Training. 
from sklearn.model_selection import train_test_split
y= dfr['Avis_efficacite_travail_confine1111']
x = dfr.drop(columns=['Avis_efficacite_travail_confine1111','Avis_efficacite_travail_confine'], axis = 1)
len (y),len(x)
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=.25, random_state=42)


std_train_x=std_scalar.fit_transform(train_x.drop(columns='genre_num'))# uniquement fit
tr=pd.DataFrame(std_train_x,columns=train_x.drop(columns='genre_num').columns)
x_train_std= pd.concat([tr.reset_index(),train_x['genre_num'].reset_index()],axis=1)


std_test_x=std_scalar.transform(test_x.drop(columns='genre_num'))# uniquement fit
tr=pd.DataFrame(std_test_x,columns=test_x.drop(columns='genre_num').columns)
x_test_std= pd.concat([tr.reset_index(),test_x['genre_num'].reset_index()],axis=1)
```

```python
# fit le modèle aux données de training
gis_search.fit(x_train_std,train_y)
```

```python
# predire la target variable dans les données de training
predict_train_y=gis_search.predict(x_train_std)
```

```python
# Accuray Score dans les données de training
accuracy_score(train_y, predict_train_y, normalize=True)
```

```python
# predire la target variable dans les données de test
predict_test_y=gis_search.predict(x_test_std)
```

```python
# Accuray Score dans les données de test
 
accuracy_score(test_y, predict_test_y, normalize=True)
```
